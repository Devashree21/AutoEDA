### <h3 style="background-color:#3f37c9;color:white;text-align: center;padding-top: 5px;padding-bottom: 5px;border-radius: 15px 50px;"><strong><centre>Loading Dataset </centre></strong></h3>

```python
#loading the dataset
from sklearn import datasets
import pandas as pd
data = datasets.load_iris()
df = pd.DataFrame(data.data,columns=data.feature_names)
df['target'] = pd.Series(data.target)
df.head()
```

### <h3 style="background-color:#3f37c9;color:white;text-align: center;padding-top: 5px;padding-bottom: 5px;border-radius: 15px 50px;"><strong><centre>EDA with PandasProfiling</centre></strong></h3>

```python
import pandas_profiling
from pandas_profiling import ProfileReport
profile = ProfileReport(df)
profile
```

### <h3 style="background-color:#3f37c9;color:white;text-align: center;padding-top: 5px;padding-bottom: 5px;border-radius: 15px 50px;"><strong><centre>Installing SweetViz 📚 </centre></strong></h3>

### <h3 style="background-color:#3f37c9;color:white;text-align: center;padding-top: 5px;padding-bottom: 5px;border-radius: 15px 50px;"><strong><centre>EDA with SweetViz </centre></strong></h3>

```python
import sweetviz as sv

#Generating SweetViz report
report = sv.analyze(df)
report.show_html("iris_EDA_report.html") # specify a name for the report
 
```

### <h3 style="background-color:#3f37c9;color:white;text-align: center;padding-top: 5px;padding-bottom: 5px;border-radius: 15px 50px;"><strong><centre>EDA with Autoviz </centre></strong></h3>

```python
#Generating AutoViz Report #this is the default command when using a file for the dataset
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()

filename = ""
sep = ","
dft = AV.AutoViz(
    '',
    sep=",",
    depVar="",
    dfte=df,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
)
```
